```python
import cv2

# Carrega a imagem
img = cv2.imread('rinoceronte.png')

# Define os valores de deslocamento na horizontal e vertical
tx = 50
ty = 100

# Define a matriz de transformação
matrix = np.float32([[1, 0, tx], [0, 1, ty]])

# Aplica a transformação à imagem
img_translada = cv2.warpAffine(img, matrix, (img.shape[1], img.shape[0]))

```

```python
from google.colab.patches import cv2_imshow

# Mostra a imagem original
cv2_imshow(img)

# Mostra a imagem transladada
cv2_imshow(img_translada)

# Espera uma tecla ser pressionada e fecha as janelas
cv2.waitKey(0)
cv2.destroyAllWindows()
```
